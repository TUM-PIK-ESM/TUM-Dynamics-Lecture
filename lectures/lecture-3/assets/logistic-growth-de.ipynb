```julia
using DifferentialEquations, Plots
```

```julia
log(u,p,t) = p*u*(1-u)

prob = ODEProblem(log, 0.5, (0.,5.), 2.3)
```

```julia
sol = solve(prob)
```

```julia
r = 2.5
plot(solve(remake(prob; u0=0., p=r)), label="x_0 = 0.")
plot!(solve(remake(prob; u0=0.1, p=r)), label="x_0 = 0.1")
plot!(solve(remake(prob; u0=0.2, p=r)), label="x_0 = 0.2")
plot!(solve(remake(prob; u0=0.3, p=r)), label="x_0 = 0.3")
plot!(solve(remake(prob; u0=0.5, p=r)), label="x_0 = 0.5")
plot!(solve(remake(prob; u0=0.7, p=r)), label="x_0 = 0.7")
plot!(solve(remake(prob; u0=0.8, p=r)), label="x_0 = 0.8")
plot!(solve(remake(prob; u0=0.9, p=r)), label="x_0 = 0.9", title="Logistic Growth (Differential Equations)", ylabel="x")
```
